## Transfer Learning

Transfer Learning is at the heart of language models like Embeddings from Language Models (ELMo) amd Bidirectional Encoder Representations from Transformer (BERT) - which can be used for any downstream task.

Every time we train a model on dataset and try to use it for a different dataset, the performance of the model deterioates. This happens as the model fails to generalize and understand basic patterns in data and even a slight difference can catch it off-guard. 

Many a times, it is very difficult to find dataset for each type of problem and won't it be great if we can utlize the model that has been trained on a similar problem for predicting values for the new type of problem.

To leverage the older model and train it for some labelled data for the new model. 

Types of Transfer Learning
1. Domain Adaptation
2. Cross-lingual Learning
3. Multi-task Learning
4. Sequential Transfer Learning

## Cross-Lingual Learning
$X_{s} != X_{t}$ 
Feature space of source and target is different. Example: specific words of stocks are different from cryptocurrency. Also, if the language is differ, there can be a complete mismatch of feature space.

## Domain Adaptation
$P_{s}(X_{s}) != P_{t}(X_{t})$
The marginal probability distribution of words is different for source and target.
It is one of the most common scenario. It can be done with either no or minimal label data for target.

Representational Approaches: try to change the undelying distribution of data by either finding features that are present in both domain or represent both data in a low-dimensional space.

1. Distribution Similarity Approaches: Prime reason behind this is to make source and target data distribution similar. A naive way to achieve similar distribution is to ignore features that do not occur in target. Major of these approaches rely on a metric of similarity calculated by available measures such as:
    - Kullback-Leiber (KL) divergence
    - Jensen-Shannon (JS) divergence
    - Wasserstein distance

For distribution similarity approaches, a common strategy is to use a representation that minimizes distance between representations of two domains, while at the same time maximizing performance on source domain data

2. Latent Feature Learning: tries to represent data in lower dimensional feature space such that it increases similarity with both source and target. The lower dimensional space can be learnt by either a factorization algorith, like Singular Value Decomposition (SVD) or a Neural Network Encoder.

$ Y_{s} != Y_{t} $ 
Labels differ for source and target. For example, Source had positive and negative labels while target has neutral as well

$ P_{s}(Y_{s}) != P_{t}(Y_{t})$
The marginal probability distribution of labels is different for source and target. Positive labels occur more than negative in training but the market crashed and real data has more negative labels than positive.

$ P_{s} (Y_{s}/X_{s}) != P_{s}(Y_{t}/X_{t})$
The condition probability distribution of labels is different. This can happen of same words can mean different or data imbalance is different for source and target.
